Plot the curve of  

$f(x)=sin(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(1, dpi=80)
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x))
plt.grid()
plt.show()
```

Plot the curve of  

$f(x)=|x^2-30|$

```python
plt.figure(1, dpi=80)
x = np.linspace(-10, 10, 100)
plt.plot(x, np.abs(x*x-30))
plt.grid()
plt.show()
```
